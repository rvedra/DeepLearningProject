Abaixo temos uma função para avaliar a proporção de outliers nos dados, utilizando a medida padrão de 1,5x o interquartil para cada uma das colunas numéricas.

```python
for c in corporate_data.columns[6:31]:

    q1 = corporate_data[c].quantile(0.25)
    q3 = corporate_data[c].quantile(0.75)
    iqr = q3 - q1 #Interquartile range
    fence_low  = q3-1.5*iqr
    fence_high = q1+1.5*iqr
    lower_out = len(corporate_data.loc[(corporate_data[c] < fence_low)  ,c])
    upper_out = len(corporate_data.loc[(corporate_data[c] > fence_high)  ,c])
    outlier_count = upper_out+lower_out
    prop_out = outlier_count/len(corporate_data)
    print(c, ": "+"{:.2%}".format(prop_out))
```

É possível identificar que há uma grande presença de outliers nos dados que precisaremos tratar em alguma etapa futura do estudo do modelo, inclusive devido à presença de assimetria nos dados.

```python
corporate_data.skew()
```

```python
df_outliers = corporate_data.copy()

for col in df_outliers.columns[6:]:
    
    q1 = df_outliers[col].quantile(0.25)
    q3 = df_outliers[col].quantile(0.75)
    iqr = q3 - q1 #Interquartile range
    fence_low  = q3-1.5*iqr
    fence_high = q1+1.5*iqr
    
    for i in range(len(df_outliers)):
        
        if corporate_data.loc[i,col] < fence_low or corporate_data.loc[i,col] > fence_high: # if Outlier
            
            df_outliers.loc[i,col] = 1
        
        else: # Not Outlier
            df_outliers.loc[i,col] = 0
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
def remove_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]
```

```python
X = remove_variance(pd.DataFrame(corporate_data[numeric_columns]), threshold=0.01)
X
```
