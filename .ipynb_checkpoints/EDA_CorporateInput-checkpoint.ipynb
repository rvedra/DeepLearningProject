Quando olhamos os dados de Rating vemos que as avaliações de extremos, AAA (menor risco de crédito), CC e C (Alto risco de crédito) e D (em default) tem pouca presença nos dados, então devemos avaliar em uma proxima etapa ignorar estes dados pois não agregariam informação relevante.

A seguir vamos verificar a presença de valores nulos e como os dados que avaliaremos como possíveis features se comportam.

```python
rating_dict = {'AAA':'Lowest Risk', 
               'AA':'Low Risk',
               'A':'Low Risk',
               'BBB':'Medium Risk', 
               'BB':'High Risk',
               'B':'High Risk',
               'CCC':'Highest Risk', 
               'CC':'Highest Risk',
               'C':'Highest Risk',
               'D':'In Default'}

corporate_data.Rating = corporate_data.Rating.map(rating_dict)
```

```python
ax = corporate_data['Rating'].value_counts().plot(kind='bar',
                                             figsize=(8,4),
                                             title="Count of Rating by Type",
                                             grid=True)
```

```python
corporate_data = corporate_data[corporate_data['Rating']!='Lowest Risk'] # filter Lowest Risk
corporate_data = corporate_data[corporate_data['Rating']!='In Default']  # filter In Default
corporate_data.reset_index(inplace = True, drop=True) # reset index
```

```python
#Checar se há dados faltantes
corporate_data.isnull().sum(axis=0)

## A princípio não há indicação de que há valores faltantes
```

```python
cols = corporate_data.iloc[:,6:].columns

for col in cols:
    plt.boxplot(corporate_data[col])
    plt.title(f'{col}')
    plt.xlabel('Valores')
    plt.show()
```

Abaixo temos uma função para avaliar a proporção de outliers nos dados, utilizando a medida padrão de 1,5x o interquartil para cada uma das colunas numéricas.

```python
for c in corporate_data.columns[6:31]:

    q1 = corporate_data[c].quantile(0.25)
    q3 = corporate_data[c].quantile(0.75)
    iqr = q3 - q1 #Interquartile range
    fence_low  = q3-1.5*iqr
    fence_high = q1+1.5*iqr
    lower_out = len(corporate_data.loc[(corporate_data[c] < fence_low)  ,c])
    upper_out = len(corporate_data.loc[(corporate_data[c] > fence_high)  ,c])
    outlier_count = upper_out+lower_out
    prop_out = outlier_count/len(corporate_data)
    print(c, ": "+"{:.2%}".format(prop_out))
```

É possível identificar que há uma grande presença de outliers nos dados que precisaremos tratar em alguma etapa futura do estudo do modelo, inclusive devido à presença de assimetria nos dados.

```python
corporate_data.skew()
```

```python
df_outliers = corporate_data.copy()

for col in df_outliers.columns[6:]:
    
    q1 = df_outliers[col].quantile(0.25)
    q3 = df_outliers[col].quantile(0.75)
    iqr = q3 - q1 #Interquartile range
    fence_low  = q3-1.5*iqr
    fence_high = q1+1.5*iqr
    
    for i in range(len(df_outliers)):
        
        if corporate_data.loc[i,col] < fence_low or corporate_data.loc[i,col] > fence_high: # if Outlier
            
            df_outliers.loc[i,col] = 1
        
        else: # Not Outlier
            df_outliers.loc[i,col] = 0
```

```python
df_outliers.head(5)
```

```python
df_outliers["total"] = df_outliers.sum(axis=1)
df_outliers.total.hist(bins = 20)
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()

for c in corporate_data.columns[6:31]:

    corporate_data[[c]] = min_max_scaler.fit_transform(corporate_data[[c]].to_numpy())*1000
    corporate_data[[c]] = corporate_data[[c]].apply(lambda x: np.log10(x+0.01))
```

```python
df_rating_no_out = corporate_data.copy()

for c in df_rating_no_out.columns[6:31]:

    q05 = df_rating_no_out[c].quantile(0.10)
    q95 = df_rating_no_out[c].quantile(0.90)
    iqr = q95 - q05 #Interquartile range
    fence_low  = q05-1.5*iqr
    fence_high = q95+1.5*iqr
    df_rating_no_out.loc[df_rating_no_out[c] > fence_high,c] = df_rating_no_out[c].quantile(0.25)
    df_rating_no_out.loc[df_rating_no_out[c] < fence_low,c] = df_rating_no_out[c].quantile(0.75)
```

```python
figure, axes = plt.subplots(nrows=8, ncols=3, figsize=(20,44))

i = 0 
j = 0

for c in df_rating_no_out.columns[6:30]:
    
    sns.boxplot(x=df_rating_no_out.Rating, y=df_rating_no_out[c], palette="Set3", ax=axes[i, j])
    
    if j == 2:
        j=0
        i+=1
    else:
        j+=1
```



A partir de agora iremos trabalhar no modelo de projeção de rating, que deverá classificar uma empresa a partir de seus indicadores finaneceiros em quatro categorias de risco, conforme vimos acima.


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import metrics


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
le = preprocessing.LabelEncoder()
le.fit(corporate_data.Sector)
corporate_data.Sector = le.transform(corporate_data.Sector) # Transformar o setor para uma variavel utilizavel
le.fit(corporate_data.Rating)
corporate_data.Rating = le.transform(corporate_data.Rating) # Transformar o Rating obtido para uma variavel utilizavel
```

```python
#Vamos separar a variavel de teste
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(corporate_data, test_size=0.2, random_state = 1234)
```

```python
X_train, y_train = df_train.iloc[:,5:31], df_train.iloc[:,0]
X_test, y_test = df_test.iloc[:,5:31], df_test.iloc[:,0]
```

**Random Forest**

```python
RF_model = RandomForestClassifier(random_state=1234)
RF_model.fit(X_train,y_train)
y_pred_RF = RF_model.predict(X_test)
Accuracy_RF = metrics.accuracy_score(y_test, y_pred_RF)
print("RF Accuracy:",Accuracy_RF)
```

**Neural Network**

```python
MLP_model = MLPClassifier(hidden_layer_sizes=(5,5,5), activation='logistic', solver='adam', max_iter=1500)
MLP_model.fit(X_train, y_train)
y_pred_MLP = MLP_model.predict(X_test)
Accuracy_MLP = metrics.accuracy_score(y_test, y_pred_MLP)
print("MLP Accuracy:",Accuracy_MLP)
```

**Linear Discriminant Analysis LDA**

```python
LDA_model = LinearDiscriminantAnalysis()
LDA_model.fit(X_train,y_train)
y_pred_LDA = LDA_model.predict(X_test)
Accuracy_LDA = metrics.accuracy_score(y_test, y_pred_LDA)
print("LDA Accuracy:",Accuracy_LDA)
```

**Logistic Regression**

```python
LR_model = LogisticRegression(random_state=1234 , multi_class='multinomial', solver='newton-cg')
LR_model = LR_model.fit(X_train, y_train)
y_pred_LR = LR_model.predict(X_test)
Accuracy_LR = metrics.accuracy_score(y_test, y_pred_LR)
print("LR Accuracy:",Accuracy_LR)
```

**Key Nearest Neightbors KNN**

```python
KNN_model = KNeighborsClassifier(n_neighbors = 3)
KNN_model.fit(X_train,y_train)
y_pred_KNN = KNN_model.predict(X_test)
Accuracy_KNN = metrics.accuracy_score(y_test, y_pred_KNN)
print("KNN Accuracy:",Accuracy_KNN)
```

```python
accuracy_list = [Accuracy_RF, Accuracy_MLP, Accuracy_LDA, Accuracy_LR, Accuracy_KNN]

model_list = ['Random Forest', 'Neural Network', 'Linear Discriminat', 'Logistic Regression', 'KNN']

df_accuracy = pd.DataFrame({'Model': model_list, 'Accuracy': accuracy_list})
```

```python
import matplotlib.ticker as mtick

order = list(df_accuracy.sort_values('Accuracy', ascending=False).Model)
df_accuracy = df_accuracy.sort_values('Accuracy', ascending=False).reset_index().drop(['index'], axis=1)

plt.figure(figsize=(12,8))
# make barplot and sort bars
x = sns.barplot(x='Model', y="Accuracy", data=df_accuracy, order = order, palette="rocket")
plt.xlabel("Model", fontsize=20)
plt.ylabel("Accuracy", fontsize=20)
plt.title("Accuracy by Model", fontsize=20)
plt.xticks(rotation=70, fontsize=12)
plt.ylim(0,1)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1))

for i in range(len(model_list)):
    plt.text(x = i, y = df_accuracy.loc[i, 'Accuracy'] + 0.05, s = str(round((df_accuracy.loc[i, 'Accuracy'])*100, 2))+'%', 
             fontsize = 14, color='black',horizontalalignment='center')

y_value=['{:,.2f}'.format(x) + '%' for x in ax.get_yticks()]
ax.set_yticklabels(y_value)

plt.tight_layout()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_RF, target_names = ['Medium Risk','Highest Risk', 'Low Risk', 'High Risk']))
```

```python
from numpy import sort
from sklearn.feature_selection import SelectFromModel

thresholds = sort(RF_model.feature_importances_)
for thresh in thresholds:
    # select features using threshold
    selection = SelectFromModel(RF_model, threshold=thresh, prefit=True)
    select_X_train = selection.transform(X_train)
    # train model
    selection_model = RandomForestClassifier()
    selection_model.fit(select_X_train, y_train)
    # eval model
    select_X_test = selection.transform(X_test)
    y_pred = selection_model.predict(select_X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = metrics.accuracy_score(y_test, predictions)
    print("Thresh=%.3f, n=%d, Accuracy: %.2f%%" % (thresh, select_X_train.shape[1], accuracy*100.0))
```
