```python
#Avaliar outliers
```

```python
for col in corporate_data.iloc[:,6:].columns:
    fmean = corporate_data[col].mean()
    fstd = corporate_data[col].std()

    # Identificar e tratar upper outliers
    upper_outliers = corporate_data.loc[np.abs(corporate_data[col] - fmean) > (3 * fstd), col]
    corporate_data.loc[np.abs(corporate_data[col] - fmean) > (3 * fstd), col] = fmean + (3 * fstd)

    # Identificar e tratar lower outliers
    lower_outliers = corporate_data.loc[np.abs(corporate_data[col] - fmean) < -(3 * fstd), col]
    corporate_data.loc[np.abs(corporate_data[col] - fmean) < -(3 * fstd), col] = -(fmean + (3 * fstd))

    # Exibir os outliers identificados
    if not upper_outliers.empty or not lower_outliers.empty:
        print(f"Outliers para a coluna {col}:")
        if not upper_outliers.empty:
            print(f"Upper outliers: {upper_outliers.tolist()}")
        if not lower_outliers.empty:
            print(f"Lower outliers: {lower_outliers.tolist()}")
        print("\n")
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
def remove_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]
```

```python
X = remove_variance(pd.DataFrame(corporate_data[numeric_columns]), threshold=0.01)
X
```
